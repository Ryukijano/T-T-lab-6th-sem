```python
#statement: Write a Python program to find the second smallest number in a list.
#modified: Write a Python program to find the fourth largest number in a list
def fourth_largest(numbers):
    if (len(numbers) < 4):
        return None
    if ((len(numbers) == 4) and (numbers[0] == numbers[1]) and (numbers[1] == numbers[2]) and (numbers[2] == numbers[3])):
        return
    duplicate = []
    unique = []
    for x in numbers:
        if x not in duplicate:
            unique.append(x)
            duplicate.append(x)
    unique.sort()
    return unique[-4]

print(fourth_largest([1, 2, -8, -2, 0, -2]))
print(fourth_largest([1, 1, 0, 0, 2, -2, -2]))
print(fourth_largest([1, 1, 1, 0, 0, 0, 2, -2, -2]))
print(fourth_largest([2,2,8,9,10,-1]))
print(fourth_largest([2,3,-3,-9,1]))
```

```python
#statement:Write a Python program to get unique values from a list.
#modified:Write a Python program to get duplicate values from a list.
my_list = []
n = int(input("Enter the number of elements in the list: "))
for i in range(0,n):
    ele = int(input("Enter the element: "))
    my_list.append(ele)
print("Original list: ", my_list)
my_set = set(my_list)
my_new_list = list(my_set)
print("New list: ", my_new_list)
```
